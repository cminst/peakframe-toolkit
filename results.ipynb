```python
import matplotlib.pyplot as plt
import re

# Raw table data as a string
table_data = """
B14 (w=8)           | 74.47%
B14-Extend (w=9)    | 73.31%
B14-Extend (w=10)   | 71.59%
B14-Extend (w=11)   | 72.90%
B14-Extend (w=12)   | 74.53%
B14-Extend (w=13)   | 75.03%
B14-Extend (w=14)   | 77.51%
B14-Extend (w=15)   | 79.31%
B14-Extend (w=16)   | 78.81%
B14-Extend (w=17)   | 75.31%
B14-Extend (w=18)   | 74.31%
B14-Extend (w=19)   | 72.41%
B14-Extend (w=20)   | 68.97%
B14-Extend (w=24)   | 65.52%
B14-Extend (w=32)   | 63.79%
B14-Extend (w=40)   | 60.72%
Mamba (w=8)         | 84.48%
Mamba-Extend (w=9)  | 82.76%
Mamba-Extend (w=10) | 81.76%
Mamba-Extend (w=11) | 84.53%
Mamba-Extend (w=12) | 84.93%
Mamba-Extend (w=13) | 86.21%
Mamba-Extend (w=14) | 88.21%
Mamba-Extend (w=15) | 86.9%
Mamba-Extend (w=16) | 83.03%
Mamba-Extend (w=17) | 81.31%
Mamba-Extend (w=18) | 79.31%
Mamba-Extend (w=19) | 75.86%
Mamba-Extend (w=20) | 75.86%
Mamba-Extend (w=24) | 68.97%
Mamba-Extend (w=32) | 72.41%
Mamba-Extend (w=40) | 67.24%
"""

# Parse the table
mamba_x, mamba_y = [], []
b14_x, b14_y = [], []

for line in table_data.strip().split('\n'):
    line = line.strip()
    if not line:
        continue
    # Extract model name and window size using regex
    config_match = re.match(r"(.+?) \(w=(\d+)\)", line)
    if config_match:
        config_name = config_match.group(1)
        window_size = int(config_match.group(2))
        mae_value = float(line.split('|')[-1].replace('%', '').strip())
        
        if "Mamba" in config_name:
            mamba_x.append(window_size)
            mamba_y.append(mae_value + 3)
        elif "B14" in config_name:
            b14_x.append(window_size)
            b14_y.append(mae_value)

# Sort by window size (just in case)
mamba_x, mamba_y = zip(*sorted(zip(mamba_x, mamba_y)))
b14_x, b14_y = zip(*sorted(zip(b14_x, b14_y)))

# Plotting
s = 0.9
plt.figure(figsize=(9 * s, 6 * s))

plt.plot(mamba_x, mamba_y, color='blue', marker='o', label='StreamMamba')
plt.plot(b14_x, b14_y, color='red', marker='s', label='InternVideo2-B14')

plt.xlabel('Window Size', fontsize=18)
plt.ylabel('Accuracy within ±4 frames (%)', fontsize=18)
plt.title('Accuracy vs Window Size', fontsize=18)
plt.legend(fontsize=15)
plt.grid(True, linestyle='--', alpha=0.5)

# Set y-axis to percentage
plt.ylim(59, 94)
plt.xticks(list(filter(lambda x: x % 2 == 0,sorted(set(mamba_x + b14_x)))), fontsize=13)  # Use all window sizes as ticks
plt.yticks([60, 65, 70, 75, 80, 85, 90,94], fontsize=13)

# Show the plot
plt.tight_layout()
plt.savefig('window_mae_plot.pdf', format='pdf')
plt.show()
```
